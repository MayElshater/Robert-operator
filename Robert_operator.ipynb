```python
import cv2
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('house.jpg')
# converting to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# remove noise
img2gray = cv2.GaussianBlur(gray,(3,3),0)
```

```python
#preparing for robert operation

kernelx = np.array( [[1, 0 ],
                             [0,-1 ]] )

kernely = np.array( [[ 0, 1 ],
                             [ -1, 0 ]] )

vertical = cv2.filter2D( img2gray,-1, kernelx )
horizontal = cv2.filter2D( img2gray,-1, kernely )

plt.figure(figsize=(20,8))
plt.subplot(221),plt.imshow(img),plt.title('Original'),plt.axis('off')
plt.subplot(222),plt.imshow(vertical),plt.title('vertical'),plt.axis('off')
plt.subplot(223),plt.imshow(horizontal),plt.title('horizontal'),plt.axis('off')
plt.subplot(224),plt.imshow(horizontal+vertical),plt.title('Robert'),plt.axis('off')
plt.show()
```
